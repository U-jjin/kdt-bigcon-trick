# 제주도 기상데이터 불러오기

```python
df0 = pd.read_csv('/content/drive/MyDrive/세미2차/제주도종관기상관측_2018010120210630.csv', encoding='cp949')
df0
```

```python
# 결측치가 가장 큰 변수 제거
df0 = df0.drop('일 최심적설(cm)', axis=1)
```

```python
# 기상관측이므로 0으로 결측값 처리
df1 = df0.fillna(0)
```

# 기온 변수간 상관관계 파악 후 변수 선택

* 평균기온&최저기온&최고기온&평균지면온도 관련 매우 높음
* 현지기압&해면기압 관련 높음
* 나머지는 다 상관성이 낮음
* 평균기온과 현지기압 선택

```python
df1.corr()
```

```python
df1 = df1.drop(['최저기온(°C)', '최고기온(°C)', '평균 지면온도(°C)', '평균 해면기압(hPa)'], axis=1)
df1
```

```python
df_고산 = df1[df1['지점명']=='고산']
df_성산 = df1[df1['지점명']=='성산']
df_제주 = df1[df1['지점명']=='제주']
df_서귀포 = df1[df1['지점명']=='서귀포']
```

# 관측소 위치에 따라 지역구간 설정
- 고산 : 제주 제주시 한경면 노을해안로 1013-70 고산기상대
- 서귀포 : 제주 서귀포시 태평로421번길 7 정방동주민센터
- 성산 : 제주 서귀포시 성산읍 신난로 12-3 성산기상대
- 제주 : 제주 제주시 일도일동 1186 제주지방기상청
-------------------------------------------------------------------
- 고산기상대 -> 제주시서, 서귀포서(=제주_서)
- 성산기상대 -> 제주시동, 서귀포동(=제주_동)
- 제주 -> 제주시_중
- 서귀포 -> 서귀포_중

```python
df_고산1 = df_고산.copy()
df_성산1 = df_성산.copy()
```

```python
df_고산['region'] = '제주시_서'
df_고산1['region'] = '서귀포시_서'
df_성산['region'] = '제주시_동'
df_성산1['region'] = '서귀포시_동'
df_제주['region'] = '제주시_중'
df_서귀포['region'] = '서귀포시_중'
```

```python
df_성산.shape
```

# 행 기준으로 합치기

```python
# 행 기준으로 일반적 merge
# 컬럼 수와 컬럼 이름이 같은 경우에만(차원이 같은 경우만)
merge1 = pd.concat([df_고산, df_고산1, df_서귀포, df_성산, df_성산1, df_제주])
```

```python
merge1
```

```python
merge2 = merge1.drop(['지점', '지점명'], axis=1)
merge2 = merge2.reset_index()
```

```python
merge2 = merge2.drop('index', axis=1)
```

```python
merge2.rename(columns={'일시':'base_date', '평균기온(°C)':'평균기온', '일강수량(mm)':'일강수량', '최대 풍속 풍향(16방위)':'최대풍속방향',
                       '평균 풍속(m/s)':'평균풍속', '평균 상대습도(%)':'평균상대습도', '평균 현지기압(hPa)':'평균현지기압'}, inplace=True)
```
